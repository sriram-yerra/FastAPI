What is await?

await is used to pause an async function until an asynchronous operation finishes ‚Äî without blocking the program.

‚ÄúWait here, but let other tasks run.‚Äù

**Why await exists**

Many operations take time:
- Database queries
- API calls
- File I/O
- Network requests
 
If Python blocked while waiting, everything would freeze.

"await" solves this by:
- Suspending the current task
- Handing control back to the event loop
- Resuming when the result is ready

What is "metadata"?

Metadata is "data about data," providing descriptive information that explains, describes, or helps manage other data, like a book's title or a photo's creation date, making it easier to find, use, and organize digital files, databases, or web content.

Examples
- For a digital photo: File name, size, date taken, camera settings, location. 
- For a document: Author, creation date, file type, last modified date. 

What is ORM model?

- The ORM model, or Object-Relational Mapping, is a programming technique that creates a bridge between object-oriented programming (OOP) languages and relational databases. 
- It allows developers to interact with the database using their programming language's objects and syntax, rather than writing raw SQL queries. 

What is a Mental model?

- A mental model is a user's internal, simplified representation or belief about how a system (software, website, hardware) works, guiding their expectations, predictions, and interactions with it. 
- For developers and designers, understanding user mental models is crucial for creating intuitive interfaces

What is Dependency Injection?

- Dependency Injection (DI) is a design pattern where an object receives its dependencies (other objects it needs) from an external source, rather than creating them itself.
- This leads to loosely coupled, more modular, and easily testable code by separating object creation from object usage. 
- Instead of a Car class making an Engine itself, the Engine is "injected" into the Car, often via its constructor or methods, allowing different Engine types to be swapped in easily.
